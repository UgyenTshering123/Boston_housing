```python
# There are 5 basic assuption of linear Regression Algorithms :
```

##### <<<<<<1. Linear Relationship between  the features and traget>>>>>>> linear b/w the features and target

##### <<<<<< 2. Little or no Multicollinearity b/w the features>>>>>>> Multicollinearity is a state of very high inter-correlation or inter-associations among the independent variables. it is therefore a type of disturbance in the data present weakens the statistical power of the regression model. Pair plots a d heatmaps (correlation Matrix) can be used for identifying highly correlated features
